### Ejercicio 2

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
from sklearn.model_selection import train_test_split
mushroom = pd.read_csv("./mushroom.csv")

# Conteo Nan y drop
countNaN = mushroom.isna().sum()
print(countNaN)
mushroom = mushroom.dropna(how="any")
countNaN = mushroom.isna().sum()
print(countNaN)

#Encoder 
encoder = OneHotEncoder(sparse_output=False)
display(mushroom)
cap = mushroom["cap-shape"].to_numpy()
gillA = mushroom["gill-attachment"].to_numpy()
gillC = mushroom["gill-color"].to_numpy()
stem = mushroom["stem-color"].to_numpy()

cap = encoder.fit_transform(cap.reshape(-1, 1))
mushroom.drop(columns=["cap-shape"], axis=1)

gillA = encoder.fit_transform(gillA.reshape(-1, 1))
mushroom.drop(columns=["gill-attachment"], axis=1)

gillC = encoder.fit_transform(gillC.reshape(-1, 1))
mushroom.drop(columns=["gill-color"], axis=1)

stem = encoder.fit_transform(stem.reshape(-1, 1))
mushroom.drop(columns=["stem-color"], axis=1)
display(mushroom)

# X, Y + stack para matriz numpy
Y = mushroom["class"].to_numpy()
X = mushroom.drop(columns=["class"], axis=1).to_numpy()
X = np.hstack((X, cap))
X = np.hstack((X, gillA))
X = np.hstack((X, gillC))
X = np.hstack((X, stem))

print(Y.shape)
print(X.shape)

# Scaler y entreno
X = StandardScaler().fit_transform(X)
print(X)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=13)
```

### Ejercicio 3

```python
from sklearn.neural_network import MLPClassifier
import time

t_per = time.time()
model = MLPClassifier(verbose=True).fit(X_train, y_train)
t_per = time.time()-t_per
```

```python
from sklearn.metrics import accuracy_score
y_proba = model.predict_proba(X_test)
y_pred = model.predict(X_test)

print(y_test)
print(y_pred)

accuracy_p = accuracy_score(y_pred, y_test)

print(accuracy_p)
print(t_per)
```

### Ejercicio 4

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=500,
                                  random_state=42,
                                  max_depth=30,
                                  min_samples_split=2,
                                  min_samples_leaf=1,
                                  max_features='sqrt',
                                  n_jobs=-1
                                  )
t_rf = time.time()
rf_model.fit(X_train, y_train)
t_rf = time.time() - t_rf
y_pred_rf = rf_model.predict(X_test)
y_pred_proba = rf_model.predict_proba(X_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(accuracy_rf)
print(t_rf)
```
